# Processamento da tabela Predict-1


```python
df = pd.read_csv('../../../data/csv/FALHAS_PREDICT-1.csv')  # Carrega o arquivo CSV em um DataFrame
df = pd.DataFrame(df)                                       # Garante que o objeto seja um DataFrame
df.head()                                                   # Exibe as primeiras linhas do DataFrame
```

```python
df.columns # Mostra as colunas presentes em meu dataset
```

```python
df = df.drop('Unnamed: 0', axis=1)  # Remove a coluna 'Unnamed: 0'
df = df.drop(0)                     # Remove a primeira linha do DataFrame
df = df.rename(columns={            # Renomeia as colunas do DataFrame
    'Unnamed: 1': 'KNR',
    'Unnamed: 2': 'MODELO',
    'Unnamed: 3': 'COR',
    'Unnamed: 4': 'MOTOR',
    'Unnamed: 5': 'ESTACAO',
    'Unnamed: 6': 'USUARIO',
    'Unnamed: 7': 'HALLE',
    'Unnamed: 8': 'FALHA',
    'Unnamed: 9': 'S_GROUP_ID',
    'Unnamed: 10': 'DATA'
})
```

```python
df['MODELO'].value_counts() # Conta a quantidade de cada tipo de diferentes modelos de carros
```

```python
df.drop(columns='MODELO', inplace=True) # Remove a coluna MODELO
```

```python
df.drop_duplicates(inplace=True) # Remove duplicadas
```

```python
df['FALHA'].value_counts() # Conta cada tipo de falha 
```

```python
df['FALHA'] = df['FALHA'].str.upper()  # Converte todos os valores da coluna 'FALHA' para maiúsculas
df[['PEÇA', 'FALHA_PEÇA']] = df['FALHA'].str.split(' ', expand=True, n=1)  # Divide 'FALHA' em 'PEÇA' e 'FALHA_PEÇA'

df['HALLE'] = df['HALLE'].str.split(' ').str[0]  # Mantém apenas o primeiro elemento após dividir 'HALLE' por espaço
```

```python
df.head(20) # Observando as novas colunas criadas
```

```python
df['PEÇA'].value_counts() # Contando os valores da coluna peça
```

```python
# Converter a coluna 'DATA' para o tipo datetime
df['DATA'] = pd.to_datetime(df['DATA'], dayfirst=True)

# Extrair o mês e o ano
df['MES_ANO'] = df['DATA'].dt.to_period('M')

# Contar os meses presentes
contagem_meses = df['MES_ANO'].value_counts().sort_index()

print(contagem_meses)
```

```python
df.shape # Mostrando a quantidade de linhas x colunas
```

```python
# Verificando os KNR's por ano
count_starts_with_2024 = df['KNR'].str.startswith('2024').sum()
print("Número de KNRs de 2024:", count_starts_with_2024)
count_starts_with_2023 = df['KNR'].str.startswith('2023').sum()
print("Número de KNRs de 2023:", count_starts_with_2023)
```

```python
# Converter a coluna S_GROUP_ID para string
df['S_GROUP_ID'] = df['S_GROUP_ID'].astype(str)

# Pivot table para HALLE
pivot_halle = df.pivot_table(index='KNR', columns='HALLE', aggfunc='size', fill_value=0)

# Pivot table para S_GROUP_ID
pivot_sgroup = df.pivot_table(index='KNR', columns='S_GROUP_ID', aggfunc='size', fill_value=0)

# Concatenar as duas pivot tables
result_df = pd.concat([pivot_halle, pivot_sgroup], axis=1)

# Renomear as colunas para diferenciar HALLE e S_GROUP_ID
result_df.columns = [f'QTD_HALLE_{col}' for col in pivot_halle.columns] + [f'QTD_SGROUP_{col}' for col in pivot_sgroup.columns]
```

```python
# Agrupar os HALLEs por S_GROUP_ID e contar as ocorrências
agrupamento_halle_sgroup = df.groupby('S_GROUP_ID')['HALLE'].value_counts().unstack(fill_value=0)

# Exibir o DataFrame resultante
agrupamento_halle_sgroup.head(20)
```

```python
result_df.head() # Resultado da tabela pivotada
```

```python
result_df.reset_index(inplace=True)  # Reseta o índice do DataFrame e transforma o índice antigo em uma coluna
```

```python
result_df.head()
```

```python
total_zp7 = (df['HALLE'] == 'ZP7').sum()  # Conta as ocorrências de 'ZP7' na coluna 'HALLE' do DataFrame original
print(f"Total de ocorrências de ZP7 no df antigo: {total_zp7}")

total_zp7_pivot = result_df['QTD_HALLE_ZP7'].sum()  # Soma os valores na coluna 'QTD_HALLE_ZP7' do DataFrame pivot
print(f"Total de ocorrências de ZP7 no df pivot: {total_zp7_pivot}")
```

```python
motor_cor_df = df.groupby('KNR')[['MOTOR', 'COR']].first().reset_index()  # Agrupa por 'KNR', pega o primeiro 'MOTOR' e 'COR', e reseta o índice
pivot_df = pd.merge(result_df, motor_cor_df, on='KNR', how='left')        # Faz um merge de 'result_df' com 'motor_cor_df' com base em 'KNR'
```

```python
pivot_df # Observando a tabela pivotada
```

```python
pivot_df['MOTOR'] = pivot_df['MOTOR'].replace(r'^\s*$', np.nan, regex=True) # Substitui valores vazios ou espaços em branco na coluna 'MOTOR' por NaN
```

```python
# Obter uma lista de todas as colunas do DataFrame
cols = pivot_df.columns.tolist()

# Definir a nova ordem das colunas
# Certifique-se de incluir 'QTD_FALHAS_SEM_HALLE' apenas uma vez, no final
new_order = ['KNR', 'MOTOR', 'COR'] + \
            [col for col in cols if col not in ('KNR', 'MOTOR', 'COR')]  # Garante que 'QTD_FALHAS_SEM_HALLE' seja incluída uma vez

# Aplicar a nova ordem ao DataFrame
pivot_df = pivot_df[new_order]
```

```python
# Verificando as colunas atuais no DataFrame
pivot_df
```

```python
# Adicionando colunas vazias se não existirem
for col in ['MOTOR', 'COR']:
    if col not in result_df.columns:
        result_df[col] = None  
```

```python
# Obtendo a lista atual de colunas
cols = result_df.columns.tolist()

# Definindo a nova ordem das colunas
new_order = ['KNR', 'MOTOR', 'COR'] + \
            [col for col in cols if col not in ('KNR', 'MOTOR', 'COR')]

# Reorganizando o DataFrame
result_df = result_df[new_order]
```
