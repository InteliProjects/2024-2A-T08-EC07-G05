# PROCESSAMENTO TABELA FALHAS

```python
df = pd.read_parquet('FALHAS_04_05_06_2024.parquet')
```

```python
df = pd.DataFrame(df)
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df.columns = df.iloc[0]
df.drop(df.index[0], inplace=True)
```

```python
columns = df.columns.tolist()
print("Columns:", columns)
```

```python
df.drop(columns='MODELO', inplace=True) # dropa pq só tem modelo t-cross
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['FALHA'].value_counts()
```

```python
df['FALHA'] = df['FALHA'].str.upper()
```

```python
df[['PEÇA', 'FALHA_PEÇA']] = df['FALHA'].str.split(' ', expand=True, n=1)
```

```python
df.head(20)
```

```python
df['HALLE'] = df['HALLE'].str.split(' ').str[0]
```

```python
count_starts_with_2024 = df['KNR'].str.startswith('2024').sum()

print("Número de KNRs de 2024:", count_starts_with_2024)
```

```python
count_starts_with_2023 = df['KNR'].str.startswith('2023').sum()

print("Número de KNRs de 2023:", count_starts_with_2023)
```

```python
df.head()
```

```python
# Create a pivot table to count occurrences of HALLE for each KNR
pivot_df = df.pivot_table(index='KNR', columns='HALLE', aggfunc='size', fill_value=0)

# Rename the columns to include the prefix 'TOTAL_FALHAS_'
pivot_df.columns = [f'QTD_FALHAS_{col}' if col.strip() else 'QTD_FALHAS_SEM_HALLE' for col in pivot_df.columns]

# Reset index to make KNR a column if needed
pivot_df.reset_index(inplace=True)
```

```python
pivot_df.head()
```

```python
total_zp7 = (df['HALLE'] == 'ZP7').sum()

print(f"Total de ocorrências de ZP7 no df antigo: {total_zp7}")
```

```python
total_zp7_pivot = (pivot_df['QTD_FALHAS_ZP7']).sum()

print(f"Total de ocorrências de ZP7 no df pivot: {total_zp7_pivot}") # dá certinho
```

```python
pivot_df.to_parquet('falha_processadas.parquet', index=False)
```

```python
# Corrected aggregation of MOTOR and COR columns from df_falhas
motor_cor_df = df.groupby('KNR')[['MOTOR', 'COR']].first().reset_index()

# # Merge the aggregated data with pivot_df
pivot_df = pd.merge(pivot_df, motor_cor_df, on='KNR', how='left')

# # Get list of all columns, then rearrange them
# cols = pivot_df.columns.tolist()

# # Move 'MOTOR' and 'COR' right after 'KNR', and 'QTD_SEM_HALLE' at the end
# new_order = ['KNR', 'MOTOR', 'COR'] + \
#             [col for col in cols if col not in ('KNR', 'MOTOR', 'COR', 'QTD_FALHAS_SEM_HALLE')] + \
#             ['QTD_FALHAS_SEM_HALLE']  # Append QTD_SEM_HALLE at the end

# pivot_df = pivot_df[new_order]
```

```python
pivot_df['MOTOR'] = pivot_df['MOTOR'].replace(r'^\s*$', np.nan, regex=True)
```

```python
pivot_df
```

```python
# Get a list of all columns in the DataFrame
cols = pivot_df.columns.tolist()

# Define the new column order
# Make sure to only include 'QTD_FALHAS_SEM_HALLE' once, at the end
new_order = ['KNR', 'MOTOR', 'COR'] + \
            [col for col in cols if col not in ('KNR', 'MOTOR', 'COR', 'QTD_FALHAS_SEM_HALLE')] + \
            ['QTD_FALHAS_SEM_HALLE']  # Ensure this is the correct column name and it's appended once

# Apply the new order to the DataFrame
pivot_df = pivot_df[new_order]
```

```python
pivot_df['TEM_FALHA_ROD'] = np.where(pivot_df['QTD_FALHAS_ROD'] > 0, 1, 0) # criando a coluna ALVO
```
